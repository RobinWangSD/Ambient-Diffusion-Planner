```python
import json
```

```python
metadata_path = '/avl-west/nuplan/file_index_pred.json'
```

```python
with open(metadata_path, 'r') as f:
    data = json.load(f)
```

```python
a = data['file_index'].keys()
```

```python
training_paths = []

if 'file_index' in data:
    for filename, file_info in data['file_index'].items():
        # Check if it's from a training directory (not 'val')
        if file_info.get('dir', '').startswith('train'):
            training_paths.append(file_info['path'])

# Sort paths for consistency
training_paths.sort()
```
